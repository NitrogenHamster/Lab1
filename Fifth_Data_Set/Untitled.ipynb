```python
import pandas as pd

file_to_load = "lp1.csv"

df = pd.read_csv(file_to_load, names = ['Failure', 'Fx' , 'Fy', 'Fz', 
                                       'Tx', 'Ty' , 'Tz', 'Interval'])
```

```python
df.head()
```

```python
#move Target Variable to last columns in dataframe
df1 = df.pop('Failure')
df['Failure'] = df1
df.head()
```

```python
train_df = df.sample(frac= 0.8, random_state = 39423899)

test_df =df.drop(train_df.index).sample(frac=1.0)
```

```python
train_df.to_csv('lp1_train.csv', index=False)
test_df.to_csv('lp1_test.csv', index=False)
```
