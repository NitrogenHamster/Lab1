```python
import pandas as pd

file_to_load = "letter-recognition.data"

df = pd.read_csv(file_to_load, names = ["letter", "x-box", "y-box",'width', 'high',
                                       'onpix', 'x-bar', 'y-bar', 'x2bar', "y2bar",
                                       'xybar', 'x2ybr', 'xy2br', 'x-ege', 'xegvy',
                                       'y-ege', 'yegvx'])
```

```python
df.head()
```

```python
#move Target Variable to last columns in dataframe
df1 = df.pop('letter')
df['letter'] = df1

#change the column type to categorical 
df.letter = pd.Categorical(df.letter)
#display the category for the data 
df['letter'] = df.letter.cat.codes

df.head()
```

```python
train_df = df.sample(frac= 0.8, random_state = 2344893)

test_df =df.drop(train_df.index).sample(frac=1.0)
```

```python
train_df.to_csv('letter-recognition_train.csv', index=False)
test_df.to_csv('letter-recognition_test.csv', index=False)
```
